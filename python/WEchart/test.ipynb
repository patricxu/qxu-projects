```python
import math
def GetHeart():
    thetas = list(map(lambda x: x/100*360, list(range(0,100))))
    return list(map(lambda theta: [5*(1 + math.sin(theta/180 * math.pi)) ,theta], thetas))

from WEchart import *
chart = Echart('极坐标双数值', '笛卡尔心形线', align='left')
chart.use(Legend(data=['line']))
chart.use(Polar())
chart.use(Tooltip(trigger='axis', axisPointer={'type':'cross'}))
chart.use(AngleAxis())
chart.use(RadiusAxis())
chart.use(Line(coordinateSystem='polar', name='line', data=GetHeart()))
chart.plot()
```
